```python
Taipei = pd.merge(Taipei_data, Taipei_temp, how='inner', on=['年', '週'])
Taipei = pd.merge(Taipei, Taipei_humidity, how='inner', on=['年', '週', '地區'])
Taipei = pd.merge(Taipei, google_trend, how='inner', on=['年', '週'])
Taipei = pd.merge(Taipei, PM25max_perweek_2008_2017, how='inner', on=['年', '週'])
Taipei = pd.merge(Taipei, Taipei_people, how='inner', on=['年', '週'])
Taipei = pd.merge(Taipei, vacation, how='left', on=['年', '週'])
Taipei = Taipei.fillna(1)


del Taipei['縣市']
del Taipei['地區']
del Taipei['min_temp']
del Taipei['max_temp']
```

```python
#making data
n_in=8
n_out=4
def series_to_supervised(add, n_in=8, n_out=4, dropnan=True):
    data = add.copy()
    year_ = data.pop('年')
    week_ = data.pop('週')
    
    for i in range(n_in, 0, -1):
        data['t-%d' %(i)] = data['total'].shift(i)
        
    for i in range(n_in, 0, -1):
        data['temp-%d' %(i)] = data['average_temp'].shift(i)
        
    for i in range(n_in, 0, -1):
        data['tensity-%d' %(i)] = data['腸病毒搜尋強度'].shift(i)
        
    for i in range(n_in, 0, -1):
        data['humidity-%d' %(i)] = data['average_humidity'].shift(i)
        
    for i in range(n_in, 0, -1):
        data['mean_polution-%d' %(i)] = data['mean_value'].shift(i)
        
    for i in range(n_in, 0, -1):
        data['child_population-%d' %(i)] = data['0_4'].shift(i)
        
    for i in range(n_in, 0, -1):
        data['vacation-%d' %(i)] = data['放假'].shift(i)
        
    for j in range(n_out):
        if j == 0:
            data['t'] = data['total']
        else:
            data['t+%d' %(j)] = data['total'].shift(-j)
    
    del data['total']
    del data['average_temp']
    del data['腸病毒搜尋強度']
    del data['average_humidity']
    del data['mean_value']
    del data['0_4']
    del data['放假']
    
    if dropnan:
        data.dropna(inplace=True)
            
    data = data.reset_index(drop=True)
            
    return data
```

```python
#take 2017 for the testing data
def select_train_test_data(new_train):
    X = new_train[:-52].iloc[:, :-n_out].values
    y_1 = new_train[:-52].iloc[:, -n_out].values
    y_2 = new_train[:-52].iloc[:, -n_out+1].values
    y_3 = new_train[:-52].iloc[:, -n_out+2].values
    y_4 = new_train[:-52].iloc[:, -n_out+3].values
    
    test_X = new_train[-52:].iloc[:, :-n_out].values
    test_y_1 = new_train[-52:].iloc[:, -n_out].values
    test_y_2 = new_train[-52:].iloc[:, -n_out+1].values
    test_y_3 = new_train[-52:].iloc[:, -n_out+2].values
    test_y_4 = new_train[-52:].iloc[:, -n_out+3].values
    
    return X, y_1, y_2, y_3, y_4, test_X, test_y_1, test_y_2, test_y_3, test_y_4
```

```python
def train_split_train_valid(X, y, test_size = 0.2):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, shuffle=True)
    return X_train, X_val, y_train, y_val
```

```python
def model_ishere(X, y, test_X, test_y, n_estimators=8000, max_depth = 20, learning_rate=0.001, gamma=0.01, eta=0.3, alpha=0.5):
    
    X_train, X_val, y_train, y_val = train_split_train_valid(X, y)
    eval_set = [(test_X, test_y), (X_val, y_val)]
    model = XGBRegressor(n_estimators=8000, max_depth = 20, learning_rate=0.001, gamma=0.01, eta=0.3, alpha=0.5)
    model.fit(X_train, y_train, early_stopping_rounds=5, eval_metric='rmse', eval_set=eval_set, verbose=True)
    
    pred_train = model.predict(X)
    
    plt.figure(figsize=(8, 6))
    plt.plot(y)
    plt.plot(pred_train)
    plt.legend('train y', 'predict y')
    plt.show()
    
    return model, pred_train
```

```python
series_to_supervised(Taipei)
```

```python
new_train = series_to_supervised(Taipei, 8, 4)
X, y_1, y_2, y_3, y_4, test_X, test_y_1, test_y_2, test_y_3, test_y_4 = select_train_test_data(new_train)
```

### week 1 model

```python
model_1, pred_train_1 = model_ishere(X, y_1, test_X, test_y_1, n_estimators=160, max_depth =10, learning_rate=0.001, gamma=0.5, eta=0.3, alpha=0.3)
pred_test_1 = model_1.predict(test_X)
```

```python
plt.plot(model_1.evals_result()['validation_0']['rmse'])
plt.plot(model_1.evals_result()['validation_1']['rmse'])
plt.legend(['test', 'val'])
plt.show()
```
