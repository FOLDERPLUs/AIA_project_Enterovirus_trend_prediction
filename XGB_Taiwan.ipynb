```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from xgboost import plot_importance
from sklearn.metrics import mean_squared_error

path = 'data_is_here/'
```

### Load Data & process

```python
data_1 = pd.read_csv(path + 'Enterovirus_Outpatient_Hospitalization.csv').iloc[:, 1:]
data_1.head()
```

```python
data_2 = pd.read_csv(path + 'Enterovirus_Emergency.csv').iloc[:, 1:]
data_2.head()
```

### Emergency & Outpatient_Hospitalization

```python
Emergency = data_2.sort_values(['年', '週'])
Emergency.head()
```

```python
Emergency_per = Emergency.groupby(['年', '週'], as_index=False)['腸病毒急診就診人次'].sum()
Emergency_per.head()
```

```python
Outpatient_Hospitalization = data_1.sort_values(['年', '週'])
Outpatient_Hospitalization.head()
```

```python
#把住院的刪掉
Outpatient = Outpatient_Hospitalization.iloc[:, [0, 1, 4, 5]][Outpatient_Hospitalization.iloc[:, [0, 1, 2, 5]]['就診類別'] == '門診']
Outpatient_per = Outpatient.groupby(['年', '週'], as_index=False)['腸病毒健保就診人次'].sum()
Outpatient_per.head()
```

```python
six_cities_temp = pd.read_csv(path + 'six_cities_temp.csv')
six_cities_temp.head()
```

```python
#取有年跟週急診的就診人數並相加
add = pd.merge(Emergency_per, Outpatient_per, how='inner')
add["total"] = add['腸病毒急診就診人次'] + add['腸病毒健保就診人次']
del add['腸病毒急診就診人次']
del add['腸病毒健保就診人次']
add = add[:509]

add.head()
```

```python
n_in=26
n_out=4
def series_to_supervised(add, n_in=26, n_out=4, dropnan=True):
    data = add.copy()
    year_ = data.pop('年')
    week_ = data.pop('週')
    
    for i in range(n_in, 0, -1):
        data['t-%d' %(i)] = data['total'].shift(i)
        
    for j in range(n_out):
        if j == 0:
            data['t'] = data['total']
        else:
            data['t+%d' %(j)] = data['total'].shift(-j)
    
    del data['total']
    
    if dropnan:
        data.dropna(inplace=True)
            
    data = data.reset_index(drop=True)
            
    return data

new_train = series_to_supervised(add, n_in, n_out)
new_train.head()
```

```python
def select_train_test_data(new_train):
    X = new_train[:-52].iloc[:, :-n_out].values
    y_1 = new_train[:-52].iloc[:, -n_out].values
    y_2 = new_train[:-52].iloc[:, -n_out+1].values
    y_3 = new_train[:-52].iloc[:, -n_out+2].values
    y_4 = new_train[:-52].iloc[:, -n_out+3].values
    
    test_X = new_train[-52:].iloc[:, :-n_out].values
    test_y_1 = new_train[-52:].iloc[:, -n_out].values
    test_y_2 = new_train[-52:].iloc[:, -n_out+1].values
    test_y_3 = new_train[-52:].iloc[:, -n_out+2].values
    test_y_4 = new_train[-52:].iloc[:, -n_out+3].values
    
    return X, y_1, y_2, y_3, y_4, test_X, test_y_1, test_y_2, test_y_3, test_y_4
```

```python
def train_split_train_valid(X, y, test_size = 0.2):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, shuffle=True)
    return X_train, X_val, y_train, y_val
```

```python
def model_ishere(X_train, X_val, y_train, y_val, test_X, test_y):
    
    eval_set = [(X_val, y_val), (test_X, test_y)]
    model = XGBRegressor(n_estimators=6500, max_depth = 15, learning_rate=0.001, gamma=0.1, eta=0.3, alpha=0.3)
    model.fit(X_train, y_train, early_stopping_rounds=5, eval_metric='rmse', eval_set=eval_set, verbose=True)
    
    return model
```

```python
new_train = series_to_supervised(add, 26, 4)
X, y_1, y_2, y_3, y_4, test_X, test_y_1, test_y_2, test_y_3, test_y_4 = select_train_test_data(new_train)

model_list = []
pred_train_list = []
pred_test_list = []
for i, j in zip([y_1, y_2, y_3, y_4], [test_y_1, test_y_2, test_y_3, test_y_4]):
    X_train, X_val, y_train, y_val = train_split_train_valid(X, i)
    model = model_ishere(X_train, X_val, y_train, y_val, test_X, j)
    pred_train = model.predict(X)
    pred_test = model.predict(test_X)
    
    plt.plot(model.evals_result()['validation_0']['rmse'])
    plt.plot(model.evals_result()['validation_1']['rmse'])
    plt.legend(['val', 'test'])
    plt.show()
    
    model_list.append(model)
    pred_train_list.append(pred_train)
    pred_test_list.append(pred_test)
```
