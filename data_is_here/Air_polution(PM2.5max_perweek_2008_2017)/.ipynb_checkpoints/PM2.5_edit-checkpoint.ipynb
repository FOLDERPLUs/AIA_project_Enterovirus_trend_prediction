### 轉時間格式，日轉週

```python
from datetime import datetime
import calendar
```

```python
df_merge.dtypes
```

```python
df_merge['date'] = pd.to_datetime(df_merge.日期)
del df_merge['日期']
df_merge.head()
```

```python
all_date = [i.isocalendar() for i in df_merge.date]
year = [x[0] for x in all_date]
week = [y[1] for y in all_date]
```

```python
year[0]
```

```python
week[0]
```

```python
df_merge['年'] = year        #把剛剛的year  week貼到dataframe上
df_merge['週'] = week
df_merge.head()
```

```python
df_PM_perweek = df_merge.groupby(['年','週'])[['mean_value']].max()     #取同週當日平均值的最大值
df_PM_perweek.head()
```

```python
pd.options.display.float_format = '{:.3f}'.format
```

```python
plt.figure(figsize=(20,12))
df_PM_perweek.mean_value.plot(legend=(False), title = 'PM2.5 per week')

plt.ylabel('PM2.5 (max) per week')
plt.xlabel('weeks')

# plt.savefig('PM2.5取同週當日平均值的最大值.png')   #已存
plt.show() 
```
