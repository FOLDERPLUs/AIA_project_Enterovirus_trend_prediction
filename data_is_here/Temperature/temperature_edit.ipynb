```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import calendar

%matplotlib inline
```

```python
temperature = pd.read_csv('temperatue.csv')
temperature.head()
```

```python
six_big_city = ['大直', '板橋', '桃園', '臺中', '高雄', '臺南']
six_data = temperature[temperature.地區.isin(six_big_city)].reset_index(drop=True)
six_data.head()
```

```python
six_data.地區.unique()
```

```python
Taipei = six_data[six_data.地區=='大直'].set_index('時間')
New_Taipei = six_data[six_data.地區=='板橋'].set_index('時間')
Taichung = six_data[six_data.地區=='臺中'].set_index('時間')
Tainan = six_data[six_data.地區=='臺南'].set_index('時間')
Kaohsiung = six_data[six_data.地區=='高雄'].set_index('時間')
Taoyuan = six_data[six_data.地區=='桃園'].set_index('時間')
```

```python
Taipei.head()
```

```python
Taipei.index = pd.to_datetime(Taipei.index)
New_Taipei.index = pd.to_datetime(New_Taipei.index)
Taichung.index = pd.to_datetime(Taichung.index)
Tainan.index = pd.to_datetime(Tainan.index)
Kaohsiung.index = pd.to_datetime(Kaohsiung.index)
Taoyuan.index = pd.to_datetime(Taoyuan.index)
```

```python
Taipei.head()
```

```python
def give_temp_year_week(region, region_name):
    ticks = region.ix[:, ['地區', 'max_temp', 'min_temp', 'average_temp']]
    max_temp_perweek = ticks.max_temp.resample('W').mean()
    min_temp_perweek = ticks.min_temp.resample('W').mean()
    average_temp_perweek = ticks.average_temp.resample('W').mean()
    name = ['max_temp', 'min_temp', 'average_temp']
    
    all_date = [i.isocalendar() for i in max_temp_perweek.index]
    year = [x[0] for x in all_date]
    week = [y[1] for y in all_date]
    complete_data = pd.DataFrame({'年':year, '週':week, '地區':region_name, name[0]:max_temp_perweek.values, 
                                  name[1]:min_temp_perweek, name[2]:average_temp_perweek})
    for i in name:
        complete_data[i] = complete_data.pop(i)
    
    complete_data.index = range(len(complete_data))
    return complete_data
```

```python
Taipei_temp = give_temp_year_week(Taipei, '臺北')
New_Taipei_temp = give_temp_year_week(New_Taipei, '新北')
Taichung_temp = give_temp_year_week(Taichung, '臺中')
Tainan_temp = give_temp_year_week(Tainan, '臺南')
Kaohsiung_temp = give_temp_year_week(Kaohsiung, '高雄')
Taoyuan_temp = give_temp_year_week(Taoyuan, '桃園')
```

```python
Taipei_temp.head()
```

```python
six_cities_temp = pd.concat([Taipei_temp, New_Taipei_temp, Taichung_temp, Tainan_temp, Kaohsiung_temp, Taoyuan_temp], axis=0).reset_index(drop=True)
#2017年的桃園47周的缺失值補台北的
six_cities_temp.ix[2502, ['max_temp', 'min_temp', 'average_temp']] = six_cities_temp.ix[412, ['max_temp', 'min_temp', 'average_temp']]
six_cities_temp.ix[2503, ['max_temp', 'min_temp', 'average_temp']] = six_cities_temp.ix[413, ['max_temp', 'min_temp', 'average_temp']]
```
